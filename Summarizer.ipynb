```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")
```

**Text** **Summarizer**

```python
while True:
    # Get user input
    user_input = input("Enter text to summarize (or type 'exit' to quit): ")

    # Exit the loop if the user types 'exit'
    if user_input.lower() == 'exit':
        print("Exiting...")
        break

    # Perform summarization
    summary = pipe(user_input, max_length=50, min_length=25, do_sample=False)

    # Print the summary
    print("Summary:", summary[0]['summary_text'])
```

**Real-time Processing :**

```python
import requests
from bs4 import BeautifulSoup

def fetch_and_extract_text(url):
    # Fetch the HTML content of the URL with a User-Agent header
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        raise Exception(f"Failed to fetch URL: {response.status_code}")

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract the main text (customize based on the website's structure)
    # For Hindustan Times, the article text is usually inside <div> or <p> tags
    main_text = ""
    for paragraph in soup.find_all('p'):
        main_text += paragraph.get_text() + " "

    return main_text.strip()

# URL to summarize
url = "https://www.hindustantimes.com/india-news/maha-kumbh-2025-ends-today-maha-shivratri-rush-at-sangam-for-final-snan-police-on-alert-in-prayagraj-10-points-101740529530803.html"

# Fetch and extract text
article_text = fetch_and_extract_text(url)
print("Extracted Text Length:", len(article_text))
```

```python
# Summarize the text
summary = pipe(article_text, max_length=130, min_length=30, do_sample=False)

# Print the summary
print("Summary:", summary[0]['summary_text'])
```
